```python
from qsp import get_even_split_errors, get_uniform_errors, deltax_eval, plot_results
from matplotlib import pyplot
from sympy.simplify.simplify import nthroot
from sympy.abc import y
import numpy
```

```python
# examples of error functions that can be used to evaluate the performance of the QSP-based mitigation
def parabola(n):
    return n**2
def linear(n):
    return n*(10**100)
def sine(x):
    return numpy.sin(x*numpy.pi)
```

```python
# plotting the initial and reduced errors using QSP-based mitigation
x, o, r, f = get_even_split_errors(20, deltax_eval)
plot_results(x, o, r, f)
pyplot.title("Error vs Sequence Length (evenly split, deltax)")
```

```python
x, o, r, f = get_even_split_errors(20, parabola)
plot_results(x, o, r, f)
pyplot.title("Error vs Sequence Length (evenly split, parabola)")
```

```python
x, o, r, f = get_even_split_errors(20, linear)
plot_results(x, o, r, f)
pyplot.title("Error vs Sequence Length (evenly split, large linear)")

```

```python
x, o, r, f = get_even_split_errors(20, sine)
plot_results(x, o, r, f)
pyplot.title("Error vs Sequence Length (evenly split, sine)")
```

```python
x, o, r, f = get_uniform_errors(0.25*numpy.pi, 20, deltax_eval)
plot_results(x, o, r, f)
pyplot.title("Error vs Sequence Length (uniform, deltax)")
```

```python
x, o, r, f = get_uniform_errors(0.25*numpy.pi, 20, parabola)
plot_results(x, o, r, f)
pyplot.title("Error vs Sequence Length (uniform, parabola)")
```

```python
x, o, r, f = get_uniform_errors(0.25*numpy.pi, 20, linear)
plot_results(x, o, r, f)
pyplot.title("Error vs Sequence Length (uniform, large linear)")
```

```python
x, o, r, f = get_uniform_errors(0.25*numpy.pi, 20, sine)
plot_results(x, o, r, f)
pyplot.title("Error vs Sequence Length (uniform, sine)")
```
